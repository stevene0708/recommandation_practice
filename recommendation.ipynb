```
#################### Predict the user's movie recommandation ####################
# Similarity machine learning practice with pearson correlation
# However, without the best KNN, the result is the same with calculating User's mean

movie_list_path = "drive/MyDrive/ml-1m/ml-1m/"

rec_file_path = "ratings.dat"
teacher_data_path = "teacher_data.txt"
teacher_pred_path = "teacher_pred.txt"

with open(movie_list_path + rec_file_path, mode='r', encoding="UTF-8") as rec:
  with open(movie_list_path + teacher_data_path, mode='r', encoding="UTF-8") as data:
    with open(movie_list_path + teacher_pred_path, mode='r', encoding="UTF-8") as data_rec:
      rec_user_ = [[]]
      movie_rec = []
      movie_teacher = []

      i = 0
      now_user = 0
      
      for n, line in enumerate(rec):
        num_user = int(line.strip().split("::")[0])
        num_user -= 1

        if (num_user == now_user):
          rec_user_[now_user].append(line.strip().split("::")[0:3])
        else:
          temp11 = num_user + 1

          if (len(rec_user_) < temp11):
            now_user = num_user
            rec_user_.append([line.strip().split("::")[0:3]])
          else:
            rec_user_[now_user].append(line.strip().split("::")[0:3])

      for line in data:
        temp1 = line.strip().split("::")[0]
        temp2 = line.strip().split(", ")[-1]

        temp3 = []
        temp3.append(temp1)
        temp3.append(temp2)

        movie_teacher.append(temp3)
        # print(movie_teacher)

      for line in data_rec:
        movie_rec.append(line.strip().split("::")[0])
        # print(movie_rec)


      user_corr = []
      
      for pos in rec_user_:
        count = 0
        rating_user = []
        rating_teacher = []
        sum_user = 0
        sum_teacher = 0

        for pos_1 in pos:
          for m in movie_teacher:
            if (pos_1[1] == m[0]):
              rating_user.append(int (pos_1[2]))
              rating_teacher.append(int (m[1]))

              sum_user += int(pos_1[2])
              sum_teacher += int(m[1])
              
              count += 1
              # print(pos_1)
              # print(m)

        if (count == 0):
          continue

        mean_user = sum_user / count
        mean_teacher = sum_teacher / count
        mean_diff_user = 0
        mean_diff_teacher = 0
        temp_user = 0
        temp_teacher = 0
        temp_diff = 0

        for x, y in zip(rating_user, rating_teacher):
          # print(x,y)
          mean_diff_user = x - mean_user
          mean_diff_teacher = y - mean_teacher

          temp_user += mean_diff_user * mean_diff_user
          temp_teacher += mean_diff_teacher * mean_diff_teacher

          temp_diff += mean_diff_user * mean_diff_teacher

        if (temp_user != 0 and temp_teacher != 0):
          r = temp_diff / ((temp_user * temp_teacher) ** (1/2))
          arr = []
          arr.append(int (pos_1[0]))
          arr.append(r)
          # print(r)
          user_corr.append(arr)

      show = 0
      
      for n in movie_rec:
        positive = 0
        negative = 0
        count_p = 0
        count_n = 0

        for j in user_corr:

          for pos1 in rec_user_:
            if (int (pos1[0][0]) != int (j[0])):
              continue
            else:
              for pos_1 in pos1:
                if (int (pos_1[1]) == int (n)):
                  if (int (j[1]) >= 0):
                    positive += (int (j[1]) * int (pos_1[2]))
                    count_p += 1
                  else:
                    negative += (int (j[1]) * int (pos_1[2]))
                    count_n += 1
                  break

        if (count_p == 0 and positive == 0):
          count_p = 1
        if (count_n == 0 and negative == 0):
          count_n = 1

        show = (positive / count_p) - (negative / count_n)
        print("Movie's Index: %s, it's Predict score: %s" %(n, show))
```
